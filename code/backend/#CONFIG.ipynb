# GPU TEST

```python
import sys
import platform

def check_pytorch():
    print("\n==================================================")
    print("VERIFICACIÓN DE PYTORCH Y GPU")
    print("==================================================")
    try:
        import torch
        print(f"✅ PyTorch instalado correctamente.")
        print(f"   Versión: {torch.__version__}")
        
        cuda_available = torch.cuda.is_available()
        print(f"\n--- Estado de CUDA (GPU) ---")
        print(f"CUDA disponible: {'SÍ' if cuda_available else 'NO'}")
        
        if cuda_available:
            device_count = torch.cuda.device_count()
            print(f"Cantidad de GPUs detectadas: {device_count}")
            
            for i in range(device_count):
                gpu_name = torch.cuda.get_device_name(i)
                gpu_capability = torch.cuda.get_device_capability(i)
                gpu_mem = torch.cuda.get_device_properties(i).total_memory / 1e9
                print(f"\n   GPU [{i}]: {gpu_name}")
                print(f"   - Capacidad de Cómputo: {gpu_capability[0]}.{gpu_capability[1]}")
                print(f"   - Memoria Total: {gpu_mem:.2f} GB")
            
            print(f"\n   Versión de CUDA compilada en PyTorch: {torch.version.cuda}")
            if torch.backends.cudnn.is_available():
                print(f"   Versión de cuDNN: {torch.backends.cudnn.version()}")
            else:
                print("   cuDNN no está disponible.")

            # Prueba de tensores en GPU
            print("\n--- Prueba Funcional Rápida ---")
            try:
                print("   Intentando mover un tensor a la GPU...")
                x = torch.rand(5, 3)
                x_gpu = x.to('cuda')
                print(f"   ✅ ÉXITO: Tensor creado en {x_gpu.device}")
                
                print("   Realizando operación matricial simple...")
                y_gpu = torch.matmul(x_gpu, x_gpu.t())
                print(f"   ✅ ÉXITO: Operación completada. Tamaño resultado: {y_gpu.shape}")
                
            except Exception as e:
                print(f"   ❌ ERROR durante la prueba funcional: {e}")
        else:
            print("\n❌ ADVERTENCIA: PyTorch no detecta ninguna GPU.")
            print("   - Si tienes una GPU NVIDIA, asegúrate de haber instalado la versión de PyTorch con CUDA.")
            print("   - Comando recomendado: pip3 install torch torchvision torchaudio --index-url https://download.pytorch.org/whl/cu118")
            
    except ImportError:
        print("❌ PyTorch no está instalado.")
        print("   Instálalo con: pip install torch torchvision")
    except Exception as e:
        print(f"❌ Error inesperado verificando PyTorch: {e}")

if __name__ == "__main__":
    print(f"Sistema Operativo: {platform.system()} {platform.release()}")
    print(f"Python Versión: {sys.version.split()[0]}")
    
    check_pytorch()
    
    print("\n==================================================")
```
